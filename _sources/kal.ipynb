### ðŸ”· Eigenvalue dan Eigenvector

Dalam aljabar linear:

- **Eigenvector** adalah vektor tak nol yang arah dan bentuknya **tidak berubah** ketika dikalikan dengan suatu matriks. Hanya panjangnya saja yang berubah.

- **Eigenvalue** adalah angka (skalar) yang menunjukkan **seberapa besar perubahan skala** yang terjadi pada eigenvector saat dikalikan dengan matriks tersebut.

Secara matematis, hubungan antara matriks $A$, eigenvector $ \mathbf{v} $, dan eigenvalue $ \lambda $ dinyatakan sebagai:

$$
A \mathbf{v} = \lambda \mathbf{v}
$$

Persamaan ini berarti bahwa ketika matriks $A$ mengalikan vektor $ \mathbf{v} $, hasilnya adalah vektor yang searah dengan $ \mathbf{v} $, dikalikan dengan skalar $ \lambda $.


# Mencari Eigenvalue dan Eigenvector Matriks â€” Penjelasan Lengkap dan Rinci

---

Diberikan matriks persegi $A$, kita ingin mencari nilai eigen $\lambda$ dan vektor eigen $\mathbf{v}$ yang memenuhi persamaan:

$$
A \mathbf{v} = \lambda \mathbf{v}
$$

dengan syarat $\mathbf{v} \neq \mathbf{0}$.

---

## 1. Definisi dan Persiapan

Persamaan di atas dapat ditulis ulang sebagai:

$$
(A - \lambda I) \mathbf{v} = \mathbf{0}
$$

Di mana:

- $I$ adalah matriks identitas dengan ukuran yang sama dengan $A$
- $\lambda$ adalah skalar (nilai eigen)
- $\mathbf{v}$ adalah vektor kolom (vektor eigen) tak nol

Agar sistem homogen ini memiliki solusi tak trivial ($\mathbf{v} \neq \mathbf{0}$), harus berlaku:

$$
\det(A - \lambda I) = 0
$$

Persamaan ini disebut **polinomial karakteristik**.

---

## 2. Langkah Mencari Eigenvalue ($\lambda$)

1. Bentuk matriks $A - \lambda I$ dengan mengurangkan $\lambda$ pada diagonal matriks $A$.

2. Hitung determinan dari $A - \lambda I$ untuk mendapatkan polinomial karakteristik, yaitu:

$$
p(\lambda) = \det(A - \lambda I)
$$

3. Cari akar-akar polinomial $p(\lambda)$ dengan menyelesaikan:

$$
p(\lambda) = 0
$$

Nilai-nilai $\lambda$ yang memenuhi adalah **eigenvalue** dari matriks $A$.

---

## 3. Langkah Mencari Eigenvector ($\mathbf{v}$)

Untuk setiap nilai eigen $\lambda$ yang ditemukan:

1. Substitusi nilai $\lambda$ ke dalam matriks $A - \lambda I$.

2. Selesaikan sistem linear homogen berikut:

$$
(A - \lambda I) \mathbf{v} = \mathbf{0}
$$

3. Carilah semua vektor $\mathbf{v} \neq \mathbf{0}$ yang memenuhi persamaan ini. Vektor-vektor ini adalah **eigenvector** yang bersesuaian dengan nilai eigen $\lambda$.

---

## 4. Contoh 1: Matriks $A$

$$
A = \begin{bmatrix} 3 & 0 \\ 0 & 3 \end{bmatrix}
$$

### a. Cari polinomial karakteristik

Bentuk matriks $A - \lambda I$:

$$
A - \lambda I = \begin{bmatrix} 3 - \lambda & 0 \\ 0 & 3 - \lambda \end{bmatrix}
$$

Hitung determinan:

$$
\det(A - \lambda I) = (3 - \lambda)(3 - \lambda) = (3 - \lambda)^2
$$

---

### b. Cari nilai eigen

Selesaikan persamaan karakteristik:

$$
(3 - \lambda)^2 = 0 \implies \lambda = 3
$$

Nilai eigen $\lambda = 3$ dengan **multiplikitas 2**.

---

### c. Cari eigenvector untuk $\lambda = 3$

Substitusi $\lambda = 3$ ke $A - \lambda I$:

$$
A - 3I = \begin{bmatrix} 0 & 0 \\ 0 & 0 \end{bmatrix}
$$

Sistem linear:

$$
\begin{bmatrix} 0 & 0 \\ 0 & 0 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$

Karena matriks nol, persamaan ini berlaku untuk **semua vektor** tak nol:

$$
\mathbf{v} = \begin{bmatrix} x \\ y \end{bmatrix}, \quad \text{dengan } (x,y) \neq (0,0)
$$

---

## 5. Contoh 2: Matriks $B$

$$
B = \begin{bmatrix} -2 & 0 \\ 0 & 4 \end{bmatrix}
$$

### a. Cari polinomial karakteristik

Bentuk matriks $B - \lambda I$:

$$
B - \lambda I = \begin{bmatrix} -2 - \lambda & 0 \\ 0 & 4 - \lambda \end{bmatrix}
$$

Hitung determinan:

$$
\det(B - \lambda I) = (-2 - \lambda)(4 - \lambda)
$$

---

### b. Cari nilai eigen

Selesaikan persamaan:

$$
(-2 - \lambda)(4 - \lambda) = 0
$$

Maka:

$$
\lambda_1 = -2, \quad \lambda_2 = 4
$$

---

### c. Cari eigenvector untuk $\lambda_1 = -2$

Substitusi ke matriks:

$$
B - (-2)I = B + 2I = \begin{bmatrix} 0 & 0 \\ 0 & 6 \end{bmatrix}
$$

Sistem linear:

$$
\begin{bmatrix} 0 & 0 \\ 0 & 6 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$

Persamaan baris kedua:

$$
6y = 0 \implies y = 0
$$

Baris pertama selalu $0=0$, tidak membatasi $x$.

Jadi, eigenvector adalah semua vektor bentuk:

$$
\mathbf{v}_1 = \begin{bmatrix} x \\ 0 \end{bmatrix}, \quad x \neq 0
$$

Misal:

$$
\mathbf{v}_1 = \begin{bmatrix} 1 \\ 0 \end{bmatrix}
$$

---

### d. Cari eigenvector untuk $\lambda_2 = 4$

Substitusi ke matriks:

$$
B - 4I = \begin{bmatrix} -6 & 0 \\ 0 & 0 \end{bmatrix}
$$

Sistem linear:

$$
\begin{bmatrix} -6 & 0 \\ 0 & 0 \end{bmatrix} \begin{bmatrix} x \\ y \end{bmatrix} = \begin{bmatrix} 0 \\ 0 \end{bmatrix}
$$

Persamaan baris pertama:

$$
-6x = 0 \implies x = 0
$$

Baris kedua selalu $0=0$, tidak membatasi $y$.

Jadi, eigenvector adalah semua vektor bentuk:

$$
\mathbf{v}_2 = \begin{bmatrix} 0 \\ y \end{bmatrix}, \quad y \neq 0
$$

Misal:

$$
\mathbf{v}_2 = \begin{bmatrix} 0 \\ 1 \end{bmatrix}
$$



```python
import numpy as np

# Matriks A
A = np.array([[3, 0],
              [0, 3]])

# Hitung eigenvalue dan eigenvector A
eigenvalues_A, eigenvectors_A = np.linalg.eig(A)

# Urutkan indeks berdasarkan eigenvalue dari besar ke kecil
sorted_indices_A = np.argsort(eigenvalues_A)[::-1]

# Urutkan eigenvalue dan eigenvector sesuai indeks
eigenvalues_A_sorted = eigenvalues_A[sorted_indices_A]
eigenvectors_A_sorted = eigenvectors_A[:, sorted_indices_A]

# Cetak hasil untuk A
print("A =", eigenvalues_A_sorted)
print(eigenvectors_A_sorted)

# Matriks B
B = np.array([[-2, 0],
              [0, 4]])

# Hitung eigenvalue dan eigenvector B
eigenvalues_B, eigenvectors_B = np.linalg.eig(B)

# Urutkan indeks berdasarkan eigenvalue dari besar ke kecil
sorted_indices_B = np.argsort(eigenvalues_B)[::-1]

# Urutkan eigenvalue dan eigenvector sesuai indeks
eigenvalues_B_sorted = eigenvalues_B[sorted_indices_B]
eigenvectors_B_sorted = eigenvectors_B[:, sorted_indices_B]

# Cetak hasil untuk B
print()
print("B =", eigenvalues_B_sorted)
print(eigenvectors_B_sorted)
```
